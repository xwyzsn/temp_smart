```python
import multiprocessing
from smart_server import SmartServer

def run_server():
    server = SmartServer()
    server.run(host="127.0.0.1", port="8000")

def run():
    process = multiprocessing.Process(target=run_server)
    process.start()
    print("Server started in a new process.")
run()
```
